Split Data into Training and Test Data

```
X_train, X_test, y_train, y_test = train_test_split(data['review'], data['sentiment'],test_size=0.2, random_state=42)
```

Build the Tokenizer

```
from tensorflow.python.keras.preprocessing.text import Tokenizer
```

```
top_words = 10000
```

```
t = Tokenizer(num_words = top_words)
```

```
t.fit_on_texts(X_train.tolist())
```

Prepare Training and Test Data

Get the word index for each of the word in the review

```
X_train = t.texts_to_sequences(X_train.tolist())
X_test = t.texts_to_sequences(X_test.tolist())
```

```
X_train[0]
```

```
len(X_train[885])
```

Pad Sequences - Important

```
from tensorflow.python.keras.preprocessing import sequence

max_review_length = 300

X_train = sequence.pad_sequences(X_train, maxlen=max_review_length, padding='post')
X_test = sequence.pad_sequences(X_test, maxlen=max_review_length, padding='post')
```

```
X_train.shape
```

Build Embedding Matrix from Pre-Trained Word2Vec

```
!pip install -q gensim
import gensim
```

```
upload = files.upload()
```

```
for fn in upload.keys():
  print('File uploaded with {name} and {length} bytes'.format(name=fn, length= len(upload[fn])))
```

```
word2vec = gensim.models.Word2Vec.load('word2vec-movie-50')
```

Embedding Size

```
word2vec.wv.vectors.shape
```

```
embedding_vector_length = word2vec.wv.vectors.shape[1]
```

Build matrix for current data

```
embedding_matrix = np.zeros((top_words+1, embedding_vector_length))
```

```
embedding_matrix
```

```
t.word_index.items()
```

```
for word, i in sorted(t.word_index.items(), key=lambda x:x[1]):
  if i > top_words:
    break
  if word in word2vec.wv.vocab:
    embedding_word = word2vec.wv[word]
    embedding_matrix[i] = embedding_word
```

```
embedding_matrix[2]
```

```
len(embedding_matrix)
```

Build the Graph

```
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Flatten, Embedding, Dense, Dropout, Average, LSTM
```

```
model = Sequential()
```

Add Embedding layer

```
model.add(Embedding(top_words+1, embedding_vector_length, 
           input_length=max_review_length,
           weights=[embedding_matrix],
          trainable=False))
```

```
model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2,))
```

```
model.add(Dense(1, activation='sigmoid'))
```

```
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'], )
```

Execute the graph

```
model.fit(X_train, y_train,
         epochs=50,
           batch_size=128,
           validation_data=(X_test, y_test),
           verbose=1)
```
